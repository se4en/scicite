```python
import jsonlines
import numpy as np
```

```python
FILE_TEST_PRED = "preds/origin/out_test.jsonl"
FILE_TEST_TRUE = "scicite/data/acl-arc/test.jsonl"

FILE_DEV_PRED = "preds/origin/out_dev.jsonl"
FILE_DEV_TRUE = "scicite/data/acl-arc/dev.jsonl"

FILE_TRAIN_PRED = "preds/origin/out_train.jsonl"
FILE_TRAIN_TRUE = "scicite/data/acl-arc/train.jsonl"

FILE_SCICITE_TEST_TRUE = "scicite/data/scicite/test.jsonl"
FILE_SCICITE_DEV_TRUE = "scicite/data/scicite/dev.jsonl"
FILE_SCICITE_TRAIN_TRUE = "scicite/data/scicite/train.jsonl"
```

```python
with jsonlines.open(FILE_TRAIN_PRED, 'r') as reader:
    for cit in reader:
        print(cit["probabilities"])
        print(cit["prediction"])
        print()
```

Labels = [ Background, Uses, CompareOrContrast, Motivation, Extends, Future ]

```python
labels_cnt = {
    'Background': 0, 
    'Uses': 0, 
    'CompareOrContrast': 0, 
    'Motivation': 0, 
    'Extends': 0, 
    'Future': 0 
}

classes = ['Background', 'Uses', 'CompareOrContrast', 'Motivation', 'Extends', 'Future']
y = []

with jsonlines.open(FILE_TRAIN_TRUE, 'r') as reader:
    for cit in reader:
        labels_cnt[cit['intent']] += 1
        y.append(classes.index(cit['intent']))
        
with jsonlines.open(FILE_DEV_TRUE, 'r') as reader:
    for cit in reader:
        labels_cnt[cit['intent']] += 1
        y.append(classes.index(cit['intent']))
        
with jsonlines.open(FILE_TEST_TRUE, 'r') as reader:
    for cit in reader:
        labels_cnt[cit['intent']] += 1
        y.append(classes.index(cit['intent']))
```
